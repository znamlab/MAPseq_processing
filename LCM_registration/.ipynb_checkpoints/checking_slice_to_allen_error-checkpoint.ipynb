```python
#reverse engineer flatfile for s019, take a random ROI and calculate amount of regions within
%matplotlib inline
from matplotlib import pyplot as plt

import numpy
import json

with open('/Users/turnerb/Desktop/code/image_reg/flat/brain1_s019_nl.flat','rb') as fp:
  buffer = fp.read()
nDims = int(buffer[0])
shape = numpy.frombuffer(buffer, dtype=numpy.dtype('>i4'), offset=1, count=2) 
data = numpy.frombuffer(buffer, dtype=numpy.dtype('>i2'), offset=9)
data = data.reshape(shape[::-1])
print(nDims,shape,data.shape,data.min(),data.max())

plt.imshow(data)
```

```python
with open('/Users/turnerb/Desktop/code/reverseEngineerFlatFile/Rainbow 2017.json') as fp:
  colors = json.load(fp)
  
R = numpy.zeros(data.shape,numpy.uint8)
G = numpy.zeros(data.shape,numpy.uint8)
B = numpy.zeros(data.shape,numpy.uint8)
for c in numpy.unique(data):
  clr = colors[c]
  idx = (data==c)
  R[idx] = clr['red']
  G[idx] = clr['green']
  B[idx] = clr['blue']
rgb = numpy.stack((R,G,B),axis=2)
```

```python
import cv2

mask = cv2.imread('/Users/turnerb/Desktop/code/image_reg/rois/s019_tube15.png')
resized_image = cv2.resize(mask, (498, 468)) 
resized_image = cv2.cvtColor(resized_image, cv2.COLOR_BGR2GRAY)
thresh = 127
resized_image1 = cv2.threshold(resized_image, thresh, 1, cv2.THRESH_BINARY)[1]
```

```python
Inew = data*resized_image1
plt.imshow(Inew)
```

```python
import pandas as pd
df=pd.DataFrame(colors)
brainregions = pd.DataFrame(Inew[Inew != 0], columns=['code'])
brainregions['region'] = " "
brainregions.set_index("code", inplace = True)
brainregions['region'] = brainregions.index.map(df['name'])
bla = brainregions['region']
```

```python
counting = bla.value_counts().rename_axis('region').reset_index(name='counts')
region =counting.region 
counts =counting.counts

fig1, fig = plt.subplots()
fig.pie(counts, labels= region)
plt.show()
```
